### Set Parameters
Required:\
`Q`: total number of queries $Q$ (`int`)\
`B`: list of $M$ beta values teachers (`double` `list`) \
`N`: number of experiments to run (`int`)\
`R_max`: maximum reward value (`int`)\
`items`: list of elements of $\mathcal{I}$ (`Item` `list`)\
`seed`: seed for random number generation (`int`)

Optional:\
`R_dict`: reward function over item set $\mathcal{I}\rightarrow\mathbb{R}^+$ (`Item:int` `dict`)

```python
# required
Q = 1000
B = [1]
N = 1
R_max = 100
items = [Item.D, Item.E, Item.F]
seed = 123

# optional (set to None for random generation)
R_dict = None
# R_dict = {
#     Item.D: 80,
#     Item.E: 30,
#     Item.F: 90
# }
```

### Run

```python
experimentID = run(B, Q, N, R_max, items, seed, R_dict, plot=True)
```

### Plot

```python
# import data
data = import_data(experimentID, N)
```

```python
### access fields as data[run_int][field_string] ###
# EX to convert single values to numbers:
    # teacher_beta_0 = data[0]['teacher_beta'].astype(np.float)
# EX to convert arrays to numbers: 
    # truth_weights_0 = [np.fromstring(s[1:-1], dtype=float, sep=' ') for s in data[0]['truth_weights']]

# plot average truth weight against query number
query_num = data[0]['query_num'].astype(np.int)

for run in range(N):
    truth_weight = [np.fromstring(s[1:-1], dtype=float, sep=' ') for s in data[run]['truth_weights']]
    avg_truth_weight = np.mean(truth_weight, axis=1)
    plt.plot(query_num, avg_truth_weight, label = 'run {}'.format(run))

plt.title("B={}".format(B[0]))
plt.xlabel('query')
plt.ylabel('average weight on ground truth')
plt.legend()
```

```python
# plot average truth weight against query number
query_num = data[0]['query_num'].astype(np.int)

for run in range(N):
    expectations = [np.fromstring(s[1:-1], dtype=float, sep=' ') for s in data[run]['expectations']]
    for i in range(len(expectations[0])):
        plt.plot(query_num, expectations[:,i], label = 'run {}'.format(run))

plt.title("B={}".format(B[0]))
plt.xlabel('query')
plt.ylabel('expectation')
plt.legend()
```
