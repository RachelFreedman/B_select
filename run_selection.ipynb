```python
import math
import random as r
import numpy as np
import scipy.stats as stats
from decimal import Decimal
from utils import *
from selection import *
from experiment_driver import run
```

### Set Parameters
Required:\
`Q`: total number of queries $Q$ (`int`)\
`B`: list of $M$ beta values for teachers (`double` `list`) \
`N`: number of experiments to run (`int`)\
`R_max`: maximum reward value (`int`)\
`items`: list of elements of $\mathcal{I}$ (`Item` `list`)\
`seed`: seed for random number generation (`int`)

Optional:\
`R_dict`: reward function over item set $\mathcal{I}\rightarrow\mathbb{R}^+$ (`Item:int` `dict`)

```python
# required
Q = 500
B = [.1]
N = 5
R_max = 100
items = [Item.A, Item.B, Item.C, Item.D, Item.E, Item.F, Item.G]
seed = 123

# optional (set to None for random generation)
R_dict = {
    Item.A: r.randrange(R_max),
    Item.B: r.randrange(R_max),
    Item.C: r.randrange(R_max),
    Item.D: r.randrange(R_max),
    Item.E: r.randrange(R_max),
    Item.F: r.randrange(R_max),
    Item.G: r.randrange(R_max)
}
```

### Run

```python
experimentID = run(B, Q, N, R_max, items, seed, R_dict, plot=False)
```

### Plot

```python
data = import_data(experimentID, N)
```

Plot the weight that the belief distributions places on the ground-truth reward values, averaged across all items. The first plot shows the absolute values of this weight, while the second shows how it changes as a result of each query.

```python
plot_average_change_in_weight_on_ground_truth(data, B)
```

Plot the expectation of the belief distribution over possible reward values for each item. If `R_dict` is provided, the true reward values will be plotted as well for comparison.
